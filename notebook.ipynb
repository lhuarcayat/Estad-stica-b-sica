```python
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl
from openpyxl.drawing.image import Image
from openpyxl.styles import Font, Alignment, PatternFill
import os
import tempfile
import uuid

def crear_excel_completo_con_graficos(dataframe):
    # Crear directorio temporal
    temp_dir = tempfile.mkdtemp()
    print(f"📁 Directorio temporal: {temp_dir}")
    
    try:
        # Crear un writer de Excel
        with pd.ExcelWriter('analisis_completo_con_graficos.xlsx', engine='openpyxl') as writer:
            
            resumen_data = []
            
            for idx, col in enumerate(dataframe.columns):
                print(f"\n🔄 Procesando columna {idx+1}/{len(dataframe.columns)}: {col}")
                
                # Calcular frecuencias
                freq_abs = dataframe[col].value_counts().sort_index()
                freq_rel = freq_abs / len(dataframe)
                freq_rel_pct = freq_rel * 100
                
                # Crear tabla para esta columna
                tabla = pd.DataFrame({
                    'Categoría': freq_abs.index,
                    'fi': freq_abs.values,
                    'hi': freq_rel.round(4).values,
                    'hi%': freq_rel_pct.round(2).values
                })
                
                # Agregar total
                total_row = pd.DataFrame({
                    'Categoría': ['TOTAL'],
                    'fi': [freq_abs.sum()],
                    'hi': [1.0000],
                    'hi%': [100.00]
                })
                tabla = pd.concat([tabla, total_row], ignore_index=True)
                
                # Crear interpretaciones
                interpretaciones = []
                cat_mas_frecuente = freq_abs.index[0]
                freq_mas_alta = freq_abs.iloc[0]
                interpretaciones.append(f"✓ f1: {freq_mas_alta} encuestados corresponden a la categoría '{cat_mas_frecuente}', siendo la más frecuente en {col}")
                
                if len(freq_abs) > 1:
                    cat_segunda = freq_abs.index[1]
                    prob_segunda = freq_rel.iloc[1]
                    interpretaciones.append(f"✓ h2: la probabilidad de encontrar la categoría '{cat_segunda}' en {col} es de {prob_segunda:.4f}")
                
                if len(freq_abs) > 2:
                    cat_tercera = freq_abs.index[2]
                    pct_tercera = freq_rel_pct.iloc[2]
                    interpretaciones.append(f"✓ h3%: el {pct_tercera:.2f}% de los encuestados corresponde a la categoría '{cat_tercera}' en {col}")
                
                total_categorias = len(freq_abs)
                interpretaciones.append(f"✓ En total se identificaron {total_categorias} categorías diferentes en {col}, con un total de {freq_abs.sum()} observaciones válidas")
                
                # Crear gráfico de pie
                plt.figure(figsize=(10, 8))
                
                if len(freq_abs) > 5:
                    legend_labels = [f'{cat} (n={count})' 
                                   for cat, count in zip(freq_abs.index, freq_abs.values)]
                    
                    wedges, texts, autotexts = plt.pie(freq_abs.values, 
                                                      autopct='%1.1f%%',
                                                      startangle=90,
                                                      colors=plt.cm.Set3(range(len(freq_abs))),
                                                      pctdistance=1.1,
                                                      textprops={'fontsize': 9})
                    
                    plt.legend(wedges, legend_labels, 
                              title="Categorías",
                              loc="center left", 
                              bbox_to_anchor=(1, 0, 0.5, 1),
                              fontsize=10)
                else:
                    labels_with_counts = [f'{cat}\n(n={count})' for cat, count in zip(freq_abs.index, freq_abs.values)]
                    plt.pie(freq_abs.values, 
                           labels=labels_with_counts, 
                           autopct='%1.1f%%', 
                           startangle=90,
                           colors=plt.cm.Set3(range(len(freq_abs))))
                
                plt.title(f'Distribución porcentual de {col}', 
                          fontsize=14, fontweight='bold', pad=20)
                plt.axis('equal')
                
                # Crear nombre único para la imagen
                unique_id = str(uuid.uuid4())[:8]
                img_filename = os.path.join(temp_dir, f'chart_{idx}_{unique_id}.png')
                
                # Guardar gráfico
                try:
                    plt.savefig(img_filename, dpi=300, bbox_inches='tight')
                    plt.close()
                    
                    # Verificar que el archivo existe
                    if os.path.exists(img_filename):
                        print(f"  ✅ Gráfico creado: {os.path.basename(img_filename)}")
                    else:
                        print(f"  ❌ Error: Archivo no encontrado después de guardar")
                        continue
                        
                except Exception as e:
                    print(f"  ❌ Error guardando gráfico: {e}")
                    plt.close()
                    continue
                
                # Nombre de hoja válido para Excel (máximo 31 caracteres)
                sheet_name = f"Col_{idx+1}_{col[:25]}"
                # Limpiar caracteres no válidos para nombres de hoja
                for char in ['[', ']', '*', '?', ':', '/', '\\']:
                    sheet_name = sheet_name.replace(char, '_')
                
                try:
                    # Escribir tabla en Excel
                    tabla.to_excel(writer, sheet_name=sheet_name, index=False, startrow=0)
                    
                    # Obtener la hoja de trabajo
                    worksheet = writer.sheets[sheet_name]
                    
                    # Estilizar la tabla
                    for row in worksheet.iter_rows(min_row=1, max_row=len(tabla)+1):
                        for cell in row:
                            cell.font = Font(name='Arial', size=10)
                            cell.alignment = Alignment(horizontal='center')
                            if cell.row == 1:  # Header
                                cell.font = Font(name='Arial', size=11, bold=True)
                                cell.fill = PatternFill(start_color='E6E6FA', end_color='E6E6FA', fill_type='solid')
                            elif cell.row == len(tabla)+1:  # Total row
                                cell.font = Font(name='Arial', size=10, bold=True)
                                cell.fill = PatternFill(start_color='F0F0F0', end_color='F0F0F0', fill_type='solid')
                    
                    # Insertar imagen del gráfico
                    if os.path.exists(img_filename):
                        try:
                            img = Image(img_filename)
                            img.width = 480
                            img.height = 360
                            worksheet.add_image(img, 'F1')
                            print(f"  ✅ Imagen insertada en hoja: {sheet_name}")
                        except Exception as e:
                            print(f"  ❌ Error insertando imagen: {e}")
                    
                    # Añadir interpretaciones
                    start_row = len(tabla) + 3
                    
                    # Título de interpretaciones
                    title_cell = worksheet.cell(row=start_row, column=1, value=f"INTERPRETACIONES - {col}")
                    title_cell.font = Font(name='Arial', size=12, bold=True)
                    title_cell.fill = PatternFill(start_color='FFE4B5', end_color='FFE4B5', fill_type='solid')
                    
                    # Añadir cada interpretación
                    for i, interp in enumerate(interpretaciones):
                        cell = worksheet.cell(row=start_row + i + 2, column=1, value=interp)
                        cell.font = Font(name='Arial', size=10)
                        cell.alignment = Alignment(horizontal='left', wrap_text=True)
                    
                    # Ajustar ancho de columnas
                    worksheet.column_dimensions['A'].width = 60
                    worksheet.column_dimensions['B'].width = 12
                    worksheet.column_dimensions['C'].width = 12
                    worksheet.column_dimensions['D'].width = 12
                    
                    print(f"  ✅ Hoja completada: {sheet_name}")
                    
                except Exception as e:
                    print(f"  ❌ Error procesando hoja: {e}")
                    continue
                
                # Agregar al resumen
                resumen_data.append({
                    'Columna': col,
                    'Total_observaciones': freq_abs.sum(),
                    'Categorias_diferentes': total_categorias,
                    'Categoria_mas_frecuente': cat_mas_frecuente,
                    'Frecuencia_maxima': freq_mas_alta,
                    'Porcentaje_maximo': round(freq_rel_pct.iloc[0], 2)
                })
            
            # Crear hoja resumen
            if resumen_data:
                print(f"\n📊 Creando hoja resumen...")
                resumen_df = pd.DataFrame(resumen_data)
                resumen_df.to_excel(writer, sheet_name='RESUMEN_GENERAL', index=False)
                
                # Estilizar hoja resumen
                resumen_ws = writer.sheets['RESUMEN_GENERAL']
                for row in resumen_ws.iter_rows(min_row=1, max_row=len(resumen_df)+1):
                    for cell in row:
                        cell.font = Font(name='Arial', size=10)
                        cell.alignment = Alignment(horizontal='center')
                        if cell.row == 1:
                            cell.font = Font(name='Arial', size=11, bold=True)
                            cell.fill = PatternFill(start_color='90EE90', end_color='90EE90', fill_type='solid')
                
                # Ajustar anchos de columna en resumen
                for column in resumen_ws.columns:
                    max_length = 0
                    column_letter = column[0].column_letter
                    for cell in column:
                        try:
                            if len(str(cell.value)) > max_length:
                                max_length = len(str(cell.value))
                        except:
                            pass
                    adjusted_width = min(max_length + 2, 50)
                    resumen_ws.column_dimensions[column_letter].width = adjusted_width
                
                print(f"  ✅ Hoja resumen completada")
    
    except Exception as e:
        print(f"❌ Error general: {e}")
    
    finally:
        # Limpiar archivos temporales
        try:
            import shutil
            shutil.rmtree(temp_dir)
            print(f"🧹 Directorio temporal limpiado")
        except:
            print(f"⚠️  No se pudo limpiar el directorio temporal: {temp_dir}")
    
    print("\n" + "="*60)
    print("✅ Archivo 'analisis_completo_con_graficos.xlsx' creado exitosamente!")
    print(f"📁 Ubicación: {os.path.abspath('analisis_completo_con_graficos.xlsx')}")
    print("\nEl archivo incluye:")
    print("- RESUMEN_GENERAL: Vista general de todas las variables")
    print("- Una hoja por variable con:")
    print("  • Tabla de frecuencias")
    print("  • Gráfico de pie chart")
    print("  • Interpretaciones detalladas")
    print(f"\nTotal de hojas creadas: {len(resumen_data) + 1 if resumen_data else 0}")
    print("="*60)

# Ejecutar
crear_excel_completo_con_graficos(df)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import re

def limpiar_nombre_archivo(nombre):
    """Limpia el nombre para usarlo como nombre de archivo"""
    # Eliminar caracteres especiales
    nombre_limpio = re.sub(r'[^\w\s-]', '', nombre)
    # Reemplazar espacios y guiones con guión bajo
    nombre_limpio = re.sub(r'[-\s]+', '_', nombre_limpio)
    return nombre_limpio.strip('_')

def guardar_pie_charts(dataframe, carpeta='graficos_pie'):
    """
    Genera y guarda gráficos de pie chart para cada columna del DataFrame
    
    Parámetros:
    - dataframe: DataFrame de pandas
    - carpeta: nombre de la carpeta donde guardar los gráficos
    """
    
    # Crear carpeta si no existe
    if not os.path.exists(carpeta):
        os.makedirs(carpeta)
        print(f"📁 Carpeta '{carpeta}' creada")
    
    print(f"🎨 Generando gráficos de pie chart...")
    print("="*60)
    
    for idx, col in enumerate(dataframe.columns):
        print(f"\n🔄 Procesando {idx+1}/{len(dataframe.columns)}: {col}")
        
        # Calcular frecuencias
        freq_abs = dataframe[col].value_counts().sort_index()
        freq_rel_pct = (freq_abs / len(dataframe)) * 100
        
        # Crear gráfico
        plt.figure(figsize=(12, 10))
        
        if len(freq_abs) > 5:
            # Para muchas categorías: usar leyenda
            legend_labels = [f'{cat} (n={count})' 
                           for cat, count in zip(freq_abs.index, freq_abs.values)]
            
            wedges, texts, autotexts = plt.pie(freq_abs.values, 
                                              autopct='%1.1f%%',
                                              startangle=90,
                                              colors=plt.cm.Set3(range(len(freq_abs))),
                                              pctdistance=1.1,
                                              textprops={'fontsize': 10})
            
            plt.legend(wedges, legend_labels, 
                      title="Categorías",
                      loc="center left", 
                      bbox_to_anchor=(1, 0, 0.5, 1),
                      fontsize=11)
        else:
            # Para pocas categorías: etiquetas directas
            labels_with_counts = [f'{cat}\n(n={count})' 
                                for cat, count in zip(freq_abs.index, freq_abs.values)]
            plt.pie(freq_abs.values, 
                   labels=labels_with_counts, 
                   autopct='%1.1f%%', 
                   startangle=90,
                   colors=plt.cm.Set3(range(len(freq_abs))),
                   textprops={'fontsize': 10})
        
        # Título del gráfico
        plt.title(f'Distribución porcentual de {col}', 
                  fontsize=16, fontweight='bold', pad=30)
        plt.axis('equal')
        
        # Agregar información adicional
        total_obs = len(dataframe)
        categorias = len(freq_abs)
        plt.figtext(0.02, 0.02, f'Total observaciones: {total_obs} | Categorías: {categorias}', 
                   fontsize=10, style='italic')
        
        # Nombre del archivo
        nombre_archivo = limpiar_nombre_archivo(col)
        if not nombre_archivo:  # Si queda vacío después de limpiar
            nombre_archivo = f"columna_{idx+1}"
        
        archivo_completo = os.path.join(carpeta, f'{nombre_archivo}.png')
        
        # Guardar gráfico
        try:
            plt.savefig(archivo_completo, dpi=300, bbox_inches='tight', 
                       facecolor='white', edgecolor='none')
            print(f"  ✅ Guardado: {archivo_completo}")
        except Exception as e:
            print(f"  ❌ Error guardando: {e}")
        
        plt.close()
    
    print("\n" + "="*60)
    print(f"✅ Proceso completado!")
    print(f"📁 Gráficos guardados en: {os.path.abspath(carpeta)}")
    print(f"📊 Total de gráficos generados: {len(dataframe.columns)}")
    print("="*60)

# Ejecutar
guardar_pie_charts(df)

# También puedes especificar una carpeta diferente
# guardar_pie_charts(df, carpeta='mis_graficos')
```
